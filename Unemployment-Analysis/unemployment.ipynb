```python
# Find the peak unemployment month for each region
peak_months = heatmap_data.idxmax(axis=1)   # Month with highest unemployment
peak_values = heatmap_data.max(axis=1)     # Value of highest unemployment

# Combine into a DataFrame
peak_analysis = pd.DataFrame({
    "Peak_Month": peak_months,
    "Peak_Unemployment_Rate(%)": peak_values.round(2)
})

print("Peak Unemployment Month & Rate by Region:")
print(peak_analysis)
```

```python
# Reset index to prepare for lineplot
trend_data = heatmap_data.reset_index().melt(
    id_vars="Region", 
    var_name="Month", 
    value_name="UnemploymentRate"
)

# Order months properly
month_order = ['January','February','March','April','May','June',
               'July','August','September','October','November','December']

plt.figure(figsize=(16,10))
sns.lineplot(
    data=trend_data,
    x="Month", y="UnemploymentRate", hue="Region",
    hue_order=heatmap_data.index,
    estimator=None, lw=2
)

plt.xticks(rotation=45)
plt.title("Unemployment Trend by Region (Month-wise)", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Scale data (important for clustering)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(heatmap_data.fillna(0))

# Apply KMeans (try 3 clusters, you can tune)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Add cluster labels to dataframe
heatmap_data['Cluster'] = clusters

# Visualize clusters
plt.figure(figsize=(12,6))
sns.scatterplot(
    x=heatmap_data.mean(axis=1),   # avg unemployment
    y=heatmap_data.std(axis=1),    # volatility
    hue=clusters, palette="Set2", s=150
)
for region in heatmap_data.index:
    plt.text(
        heatmap_data.mean(axis=1)[region], 
        heatmap_data.std(axis=1)[region], 
        region, fontsize=9
    )

plt.xlabel("Average Unemployment Rate (%)")
plt.ylabel("Volatility (Std. Dev.)")
plt.title("Regional Clustering of Unemployment Patterns")
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Pick one region (Andhra Pradesh example)
region_data = data[data['Region'] == 'Andhra Pradesh'][['Date', 'Estimated_Unemployment_Rate_(%)']]

# Rename for convenience
region_data = region_data.rename(columns={'Estimated_Unemployment_Rate_(%)': 'UnemploymentRate'})

# Ensure datetime format
region_data['Date'] = pd.to_datetime(region_data['Date'])
region_data.set_index('Date', inplace=True)

# Train ARIMA model
model = ARIMA(region_data['UnemploymentRate'], order=(1,1,1))  # (p,d,q)
model_fit = model.fit()

# Forecast next 12 months
forecast = model_fit.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Plot results
plt.figure(figsize=(12,6))
plt.plot(region_data.index, region_data['UnemploymentRate'], label='Observed')
plt.plot(forecast_mean.index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='pink', alpha=0.3)

plt.title("ARIMA Forecast of Unemployment Rate (Andhra Pradesh)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Pick Andhra Pradesh data
region_data = data[data['Region'] == 'Andhra Pradesh'][['Date', 'Estimated_Unemployment_Rate_(%)']]

# Ensure datetime
region_data['Date'] = pd.to_datetime(region_data['Date'])
region_data = region_data.set_index('Date')

# Fit ARIMA model
model = ARIMA(region_data, order=(1,1,1))
model_fit = model.fit()

# Forecast next 6 months
forecast = model_fit.get_forecast(steps=6)
forecast_summary = forecast.summary_frame()

print("Forecast for Andhra Pradesh (next 6 months):")
print(forecast_summary)
```

```python
import matplotlib.pyplot as plt

# Plot historical data
plt.figure(figsize=(10,5))
plt.plot(region_data.index, region_data['Estimated_Unemployment_Rate_(%)'], 
         label='Historical', color='blue')

# Plot forecast mean
forecast_index = pd.date_range(start=region_data.index[-1] + pd.DateOffset(1), 
                               periods=6, freq='M')
plt.plot(forecast_index, forecast_summary['mean'], label='Forecast', color='red')

# Plot confidence intervals
plt.fill_between(forecast_index, 
                 forecast_summary['mean_ci_lower'], 
                 forecast_summary['mean_ci_upper'], 
                 color='pink', alpha=0.3, label='Confidence Interval')

plt.title("Andhra Pradesh - Unemployment Rate Forecast (Next 6 Months)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Dictionary to store forecasts
all_forecasts = {}

# Loop through each region
for region in data['Region'].unique():
    try:
        # Filter region data
        region_data = data[data['Region'] == region][['Date', 'Estimated_Unemployment_Rate_(%)']]
        region_data['Date'] = pd.to_datetime(region_data['Date'])
        region_data = region_data.set_index('Date').asfreq('M')  # ensure monthly
        
        # Drop missing values
        region_data = region_data.dropna()
        
        # Skip if too little data
        if len(region_data) < 12:
            continue
        
        # Fit ARIMA model
        model = ARIMA(region_data['Estimated_Unemployment_Rate_(%)'], order=(1,1,1))
        model_fit = model.fit()
        
        # Forecast next 6 months
        forecast = model_fit.get_forecast(steps=6)
        forecast_df = forecast.summary_frame()
        
        # Store results
        all_forecasts[region] = forecast_df[['mean', 'mean_ci_lower', 'mean_ci_upper']]
    
    except Exception as e:
        print(f" Skipped {region} due to error: {e}")

#  Example: print Bihar's forecast
print("\n Bihar Forecast (next 6 months):")
print(all_forecasts['Bihar'].head(6))
```

```python
import matplotlib.pyplot as plt

# Choose some regions to visualize
regions_to_plot = ["Andhra Pradesh", "Bihar", "Assam", "Delhi"]

plt.figure(figsize=(12,8))

for region in regions_to_plot:
    if region not in all_forecasts:
        continue
    
    # Get historical data
    region_data = data[data['Region'] == region][['Date', 'Estimated_Unemployment_Rate_(%)']]
    region_data['Date'] = pd.to_datetime(region_data['Date'])
    region_data = region_data.set_index('Date').asfreq('M')
    
    # Plot history
    plt.plot(region_data.index, region_data['Estimated_Unemployment_Rate_(%)'], label=f"{region} (History)")
    
    # Forecast
    forecast_df = all_forecasts[region]
    forecast_index = pd.date_range(start=region_data.index[-1] + pd.offsets.MonthEnd(), periods=6, freq='M')
    
    plt.plot(forecast_index, forecast_df['mean'], '--', label=f"{region} (Forecast)")
    plt.fill_between(forecast_index, forecast_df['mean_ci_lower'], forecast_df['mean_ci_upper'], alpha=0.2)

plt.title("Unemployment Rate Forecasts (Next 6 Months)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.show()
```
